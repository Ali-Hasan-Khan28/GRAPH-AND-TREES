```python
class adjoint:
    def __init__(self, data):
        self.vertex = data
        self.next = None


class Graph:
    def __init__(self, vertices):
        self.v = vertices
        self.graph = [None]*self.v

    def add_edge(self, source, destination):
        node = adjoint(destination)
        node.next = self.graph[source]
        self.graph[source] = node
        node = adjoint(source)
        node.next = self.graph[destination]
        self.graph[destination] = node

    # Function to print the graph
    def print_graph(self):
        for i in range(self.v):
            print("Adjacency list of vertex {}\n head".format(i), end="")
            temp = self.graph[i]
            print(temp)
            while temp:
                print(" -> {}".format(temp.vertex), end="")
                temp = temp.next
            print(" \n")

    def BFS(self, s):

        visited = [False] * (len(self.graph))

        queue = []

        queue.append(s)
        visited[s] = True

        while queue:

            s = queue.pop(0)
            print(s, end=" ")

            list1 = self.graph[s]
            while list1:
                if visited[list1.vertex] == False:
                    queue.append(list1.vertex)
                    visited[list1.vertex] = True
                list1=list1.next


    def DFS(self,s):
        visit =set()
        self.revursive_dfs(s,visit)

    def revursive_dfs(self, s,visited):
        visited.add(s)
        print(s,end=" ")

        list2=self.graph[s]
        while list2:
            if list2.vertex not in visited:
                self.revursive_dfs(list2.vertex,visited)
            list2=list2.next

if __name__ == "__main__":
    V = 5
    graph = Graph(V)
    graph.add_edge(0, 1)
    graph.add_edge(2, 4)
    graph.add_edge(0, 2)
    graph.add_edge(1, 2)
    graph.add_edge(2, 0)
    graph.add_edge(2, 3)
    graph.add_edge(3, 3)

    # graph.print_graph()
    print("Depth First Search ")
    graph.DFS(2)
    print("\nBreadth First Search")
    graph.BFS(2)
```
